```python
from datasets import load_dataset

ds_gender = load_dataset("Pengrun/PropInfer_dataset", name="gender")
```

```python
ds_gender
```

```python
# split the dataset into two part: train_df and aux_df 
# train_df is used to train the target model and aux_df is used by the adversary to train shadow model
df = ds_gender['train'].to_pandas()
shuffled_df = df.sample(n=len(df), random_state= 0)
shuffled_df = shuffled_df.reset_index(drop=True)
train_df = shuffled_df.head(15000)
aux_df = shuffled_df.tail(len(df) - 15000)
```

```python
train_df.head()
```

```python
aux_df.head()
```

```python
#  Create fine-tuning dataset with different target ratio

train_female = train_df[train_df['gender'] == '1. female'].sample(n=3250, random_state = 50).index
train_male = train_df[train_df['gender'] == '2. male'].sample(n=3250, random_state = 50).index

df = train_df[(train_df.index.isin(train_female)) | (train_df.index.isin(train_male))]
df.head()
```

```python
from datasets import Dataset
df = df.drop('gender', axis = 1)
df = df.reset_index(drop = True)
dataset = Dataset.from_pandas(df, split='train')
dataset
```

```python
dataset.to_json("save_directory", lines=True)
```
